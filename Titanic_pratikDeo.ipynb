```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
sns.catplot('Sex', data = titanic_df)
```

```python
def male_female_child(passenger):
    # Take the Age and Sex
    age,sex = passenger
    # Compare the age, otherwise leave the sex
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
titanic_df['person'] = titanic_df[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
titanic_df[0:10]
```

```python
titanic_df['Age'].hist(bins=70)
```

```python
# Another way to visualize the data is to use FacetGrid to plot multiple kedplots on one plot

# Set the figure equal to a facetgrid with the pandas dataframe as its data source, set the hue, and change the aspect ratio.
fig = sns.FacetGrid(titanic_df, hue="Sex",aspect=4)
# Next use map to plot all the possible kdeplots for the 'Age' column by the hue choice
fig.map(sns.kdeplot,'Age',shade= True)
# set the max limit to the oldest passenger
oldest = titanic_df['Age'].max()
#since we know no one can be of negative age, set the lower limit at 0
fig.set(xlim=(0,oldest))
#now add a legend
fig.add_legend()
```

```python
# now we could have done similar plot for the 'person' column

fig = sns.FacetGrid(titanic_df, hue="person", aspect = 4)
fig.map(sns.kdeplot, 'Age', shade = True)
oldest = titanic_df['Age'].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```

```python
#doing the above same thing for class by changing hue 
fig = sns.FacetGrid(titanic_df, hue="Pclass",aspect=4)
fig.map(sns.kdeplot,'Age',shade= True)
oldest = titanic_df['Age'].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```

```python
#now lets drop the nan values
titanic_deck = titanic_df['Cabin'].dropna()
```

```python
titanic_deck.head()
```

```python
# So let's grab that letter for the deck level with a simple for loop

# Set empty list
levels = []

# Loop to grab first letter
for level in titanic_deck:
    levels.append(level[0])    

# Reset DataFrame and use factor plot
cabin_df = DataFrame(levels)
cabin_df.columns = ['Cabin']
sns.catplot('Cabin',data=cabin_df,palette='winter_d')
```

```python
#we have a 'T' deck value there which doesn't make sense, we can drop it out with the following code:

#remake cabin_d as 
cabin_df = cabin_df[cabin_df.Cabin != 'T']
sns.catplot('Cabin', data=cabin_df, palette = 'summer')
```

```python
#adding the parent/child with sibsp column
titanic_df['Alone'] =  titanic_df.Parch + titanic_df.SibSp
titanic_df['Alone']
```

```python
#if it is > 0 or equal to 0 then status will be alone 
titanic_df['Alone'].loc[titanic_df['Alone']>0] = 'With Family'
titanic_df['Alone'].loc[titanic_df['Alone'] == 0] = 'Alone'
```

```python
titanic_df.head()
```

```python
sns.catplot('Alone',data=titanic_df,palette='Blues')
```

```python
#lets see what factors helped someone survive the crash 

titanic_df["Survivor"] = titanic_df.Survived.map({0: "no", 1: "yes"})

#survived vs died 
sns.catplot('Survivor', data=titanic_df, palette='Set1')
```

```python
sns.catplot('Pclass','Survived', data=titanic_df)
```

```python
#Look like survival rates for the 3rd class are substantially lower! 
#But maybe this effect is being caused by the large amount of men in t
#he 3rd class in combination with the women and children first policy. 
#Let's use 'hue' to get a clearer picture on this.
```

```python
sns.catplot('Pclass','Survived', hue ='person', data=titanic_df)
```

```python
#the results show the older the person the more chances that they did not survive
sns.lmplot('Age','Survived', hue ='Pclass', data=titanic_df, palette = 'winter')
```

```python
# Let's use a linear plot on age versus survival using hue for class seperation
generations = [10,20,40,60,80]
sns.lmplot('Age','Survived', hue = 'Pclass', data=titanic_df, palette = 'winter', x_bins= generations)
```
